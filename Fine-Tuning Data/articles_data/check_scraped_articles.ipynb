### Ensure all the articles do not compare the duplicates and share the same format

```python
import pandas as pd
```

### National Geographic

```python
df_117 = pd.read_csv('117_national_geographic_articles.csv')
df_117 = df_117[['source', 'topic', 'link', 'title', 'content']]
df_117
```

```python
df_107 = pd.read_csv('National_geophaphic_final_processed.csv')
df_107['title'] = ''
df_107 = df_107[['Source', 'subtopic', 'link', 'title', 'content']]
df_107.columns = ['source', 'topic', 'link', 'title', 'content']
df_107
```

```python
national_geograhic = pd.concat([df_117, df_107])
national_geograhic = national_geograhic.drop_duplicates(subset=['link'], keep='first')
national_geograhic.index = range(len(national_geograhic))
national_geograhic
```
